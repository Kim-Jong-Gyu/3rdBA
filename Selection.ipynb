```python
import pandas as pd
import numpy as np

df_before=pd.read_csv('/Users/kimjong-gyu/Desktop/github/3rdBA/3rdBA/BA_TeamProject/before_selection.csv')
```

```python
df_before=df_before.drop(['action_sum'],axis=1)
```

```python
# affiliate_channel , affliate_provider , first_affiliate_tracked , signup_app ,first_device_type ,signup_method
# 위에꺼는 초기 접속에 관련된 변수 이기때문에 우리 타켓 벨류는 처음 사용 하고 예약 날짜와의 차이기 때문에 연관성이 없어보인다.
df_before=df_before.drop(['affiliate_channel','affiliate_provider','first_affiliate_tracked','signup_app','first_device_type'],axis=1)
```

```python
df_before=df_before.drop(['signup_method'],axis=1)
```

```python
# timestamp_first_active, tfa_date 는 우리 타켓 밸류를 만들기 위해 사용한 함수기 때문에 없앤다.
df_before=df_before.drop(['timestamp_first_active','tfa_date'], axis=1)
```

```python
df_before.head()
```

```python
df_before.language_levenshtein_distance.describe()
```

```python
# language_levenshtein_distance , distance_km , lat_destination , lng_destination , country_destination 
# 위에 변수들은 destination_km가 있으므로 필요가 없고 위도 경도는 단지 그 지역의 위치만 알려주기 때문에 필요가 없다고 판단했다.
df_before=df_before.drop(['language_levenshtein_distance','distance_km','lat_destination','lng_destination','country_destination'], axis=1)
```

```python
# action_type , action_detail , action 는 action_count를 구하기 위한 변수라 필요 없다고 판단 .
df_before=df_before.drop(['action_type','action_detail','action'],axis=1)
```

```python
# cat_duration 이 타켓 벨류 
# duration 값이 cat_duration 을 카테고리컬 데이터로 만든거니 필요가 없다고 판단 .
df_before=df_before.drop(['duration'],axis=1)
```

```python
# dummy 데이터 생성 
df_before.corr()
```

```python
df_before.head()
```

```python
df_before=df_before.drop(['language'], axis=1)
```

```python
df_before.head()
```

```python
catvar=['gender','first_browser','device_type']
for c in catvar:
    temp=pd.get_dummies(df_before[c],prefix=c, drop_first=True)
    df_before=pd.concat((df_before,temp),axis=1)
    
df_before=df_before.drop(catvar,axis=1)
```
